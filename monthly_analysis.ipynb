```python
import pandas as pd
from config import pw
```

```python
# Read the CSV file into a DataFrame
csv_path = 'Resources/kingcounty/seattle_sales_cleaned_data.csv'  # Replace with the actual path
df = pd.read_csv(csv_path)
df = pd.DataFrame(df)
print(df.head())
```

```python
# add column for year to monthly data
df['month'] = pd.DatetimeIndex(df['sale_date']).month
df.head()
```

```python

columns_to_drop = ['view_rainier', 'view_olympics', 'view_olympics', 'view_territorial', 'view_skyline', 'view_sound', 'view_lakewash', 'view_otherwater']
df_dropped_multiple = df.drop(columns=columns_to_drop)
df_dropped_multiple.head()
```

```python
# Get a list of column names
column_names = df_dropped_multiple.columns.tolist()
print(column_names)
```

```python
columns_to_drop = ['zoning','grade', 'fbsmt_grade', 'home_condition', 'stories', 'beds', 'bath_full', 'bath_3qtr', 'bath_half', 'garb_sqft', 'gara_sqft', 'wfnt', 'golf', 'greenbelt', 'noise_traffic', 'view_cascades']
df_cleaned = df_dropped_multiple.drop(columns=columns_to_drop)
df_cleaned.head()
```

```python
df_cleaned["sale_price"].value_counts
```

```python
df_cleaned['sale_date'].value_counts
```

```python
# Convert 'sale_date' to datetime
df_cleaned['sale_date'] = pd.to_datetime(df['sale_date'])

# Group by 'year' and 'month' and calculate average 'sale_price'
average_sale_price_per_month = df_cleaned.groupby(['year', 'month'])['sale_price'].mean()
# Group by 'year' and 'month' and calculate median 'sale_price'
median_sale_price_per_month = df_cleaned.groupby(['year', 'month'])['sale_price'].median()
# Group by 'year' and 'month' and calculate total sales using 'sum()'
total_sales_per_month = df_cleaned.groupby(['year', 'month'])['sale_price'].nunique()

print(total_sales_per_month)
#print(average_sale_price_per_month)
#print(median_sale_price_per_month)
```

```python
# Create a new DataFrame to store the results
results_df_monthly = pd.DataFrame({
    'Years': total_sales_per_month.index.get_level_values('year'),
    'Months': total_sales_per_month.index.get_level_values('month'),
    'Average_Sale_Price': average_sale_price_per_month.values,
    'Median_Sale_Price': median_sale_price_per_month.values,
    'Total_Houses_sold': total_sales_per_month.values
})

print(results_df_monthly)
```

```python
csv_path = 'Resources/kingcounty/seattle_monthly_data.csv' 
# Write the DataFrame to a CSV file
results_df_monthly.to_csv(csv_path, index=False)

```
